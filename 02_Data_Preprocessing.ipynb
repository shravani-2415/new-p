```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
```

```python
df.info()
```

```python
df.fillna('', inplace = True)
```

```python
df.loc[0, 'genres']
```

```python
import ast
def get_names(text):
    y = ast.literal_eval(text)
    words = []
    for a in y:
        words.append(a['name'].replace(' ', '-'))
    text = ','.join(words).lower()
    return text
```

```python
df['genres'] = df['genres'].apply(get_names)
df['keywords'] = df['keywords'].apply(get_names)
df['production_companies'] = df['production_companies'].apply(get_names)
df['cast'] = df['cast'].apply(get_names)
df['crew'] = df['crew'].apply(get_names)
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
print("success")
```

```python
from nlp_functions import nlp_pipeline
df['overview'] = nlp_pipeline(df['overview'])
df['tagline'] = nlp_pipeline(df['tagline'])
```

```python
df['overview'] = df['overview'].str.replace(" ", ",")
df['tagline'] = df['tagline'].str.replace(" ", ",")
```

```python
df['tags'] = df['genres'] + "," + df['keywords'] + "," + df['overview'] + "," + df['production_companies'] + ","  + df['tagline'] + ","  + df['cast'] + ","  + df['crew']
```

```python
df = df[['id', 'title', 'tags']]
```

```python
def get_unique_values(text):
    return ','.join(list(set(text.split(','))))
for i in df.index:
    df.loc[i, 'tags'] = get_unique_values(df.loc[i, 'tags'])
```

```python
df.to_csv("final_data.csv", index = False)
```
